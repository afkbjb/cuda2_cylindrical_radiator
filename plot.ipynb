```python
import matplotlib.pyplot as plt

n = [64, 128, 256, 512, 1024]
speedup = [3.45, 10.48, 38.98, 144.80, 261.44]

plt.figure(figsize=(8,6))
plt.plot(n, speedup, marker='o')
plt.xlabel('Matrix size (n = m)')
plt.ylabel('Speedup (CPU time / GPU time)')
plt.title('Speedup vs Matrix Size')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

n = [64, 128, 256, 512, 1024]
speedup_single = [3.45, 10.48, 38.98, 144.80, 261.44]
speedup_double = [2.14, 8.55, 29.33, 70.63, 164.78]

plt.figure(figsize=(8,6))
plt.plot(n, speedup_single, marker='o', label='Single Precision')
plt.plot(n, speedup_double, marker='s', label='Double Precision')
plt.xlabel('Matrix size (n = m)')
plt.ylabel('Speedup (CPU time / GPU time)')
plt.title('Speedup vs Matrix Size')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
